## Import data

```python
import pandas as pd

df1 = pd.read_csv('data/9737_electricity_egauge_hours_2014.csv',
#                 index_col= 'localhour',
                 header='infer',
                 sep=',')
```

```python
df1.dropna(axis=1,how='all');
# drop the columns where all values are null
```

```python
df1 = df1.drop('dataid', axis=1)
```

```python
df1.head()
```

```python
df1['electricity_consumption'] = df1.sum(axis=1)
```

```python
df_1 = df1['electricity_consumption'].copy()
```

```python
df_1.head()
```

```python
import pandas as pd

df2 = pd.read_csv('data/weather_hours_2014.csv',
#                 index_col='localhour',
                 header='infer',
                 sep=',')
```

```python
df2.dropna(axis=1,how='all');
# drop the columns where all values are null
```

```python
df2.median();
df_2 = df2.fillna(df2.median())
```

```python
df_2.head()
```

```python
df = pd.concat([df_2, df_1], axis=1)
# concatenate two dataframes
```

```python
df.fillna(0, inplace=True)
# set the null value to zero
```

```python
#df.fillna(df.median())
```

```python
df.head()
```

## Feature engineering

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style='darkgrid', context='notebook')
cols = ['temperature', 'humidity', 'dew_point','visibility','pressure', 'apparent_temperature',
        'wind_speed', 'cloud_cover', 'wind_bearing',
        'electricity_consumption']

sns.pairplot(df[cols], size=15)
plt.tight_layout()
plt.savefig('./figures/scatter.png', dpi=300)
# plt.show()
```

```python
import numpy as np

cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 6},
                 yticklabels=cols,
                 xticklabels=cols)

# plt.tight_layout()
plt.savefig('./figures/corr_mat.png', dpi=300)
plt.show()
```
